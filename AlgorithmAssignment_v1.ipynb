### Calculate Distance

```python
#store calculated distances
distanceList = [0]
distance1List = [0]
distance2List = [0]

for i in range(len(fullData['Latitude-Origin'])):
    #assign latitude and longitude of the points
    origin_Lat = fullData['Latitude-Origin'][i]
    origin_Long = fullData['Longitude-Origin'][i]
    dh_Lat = fullData['Latitude-DH'][i]
    dh_Long = fullData['Longitude-DH'][i]
    dest_Lat = fullData['Latitude-Destination'][i]
    dest_Long = fullData['Longitude-Destination'][i]
    
    #initialize the coordinate
    origin = (origin_Lat, origin_Long)
    del_hub = (dh_Lat, dh_Long)
    destination = (dest_Lat, dest_Long)
    
    #calculate distance
    distance1 = gmaps.distance_matrix(origin, del_hub, mode='walking')["rows"][0]["elements"][0]["distance"]["value"]
    distance2 = gmaps.distance_matrix(del_hub, destination, mode='walking')["rows"][0]["elements"][0]["distance"]["value"]
    total_distance = distance1 + distance2
    
    #add total distance in the list
    distanceList.append(total_distance)
    distance1List.append(distance1)
    distance2List.append(distance2)
    
#add distance column
fullData['Distance1'] = distance1List
fullData['Distance2'] = distance2List
fullData['Total Distance'] = distanceList


#write to csv
fullData.to_csv('calculated_distances.csv', sep=';', index=None, header= ['Customer','Origin','Latitude-Origin','Longitude-Origin','Delivery Hub','Latitude-DH','Longitude-DH','Destination','Latitude-Destination','Longitude-Destination','Distance1', 'Distance2', 'Total Distance'])
```
