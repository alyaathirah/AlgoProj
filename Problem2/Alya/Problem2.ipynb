### **3. Compare words with the positive, negative and neutral English words using a String-Matching algorithm.**

```python
def calcRatio(arr):  # positive ratio out of 5(put array)
    total = 0
    posiNum = 0
    negNum = 0
    for i in range(len(arr)):
        posiNum += len(arr[i].positive)
        negNum += len(arr[i].negative)

    total = posiNum + negNum

    star = posiNum * 1.0 / total * 5
    return star


if __name__ == '__main__':
    print("CITYLINK")
    CityLink = []
    for i in range(len(citylink)):
        CityLink.append(Website.Website(citylink[i]))
        CityLink[i].read()
        CityLink[i].PosiNeg()
        CityLink[i].Print()
    print(calcRatio(CityLink), "star")
    print()

    print("POSLAJU")
    Poslaju = []
    for i in range(len(poslaju)):
        Poslaju.append(Website.Website(poslaju[i]))
        Poslaju[i].read()
        Poslaju[i].PosiNeg()
        Poslaju[i].Print()
    print(calcRatio(Poslaju), "star")
    print()

    print("GDEX")
    GDEX = []
    for i in range(len(poslaju)):
        GDEX.append(Website.Website(gdex[i]))
        GDEX[i].read()
        GDEX[i].PosiNeg()
        GDEX[i].Print()
    print(calcRatio(GDEX), "star")
    print()

    print("J&T")
    JT = []
    for i in range(len(poslaju)):
        JT.append(Website.Website(jt[i]))
        JT[i].read()
        JT[i].PosiNeg()
        JT[i].Print()
    print(calcRatio(JT), "star")
    print()

    print("DHL")
    DHL = []
    for i in range(len(poslaju)):
        DHL.append(Website.Website(dhl[i]))
        DHL[i].read()
        DHL[i].PosiNeg()
        DHL[i].Print()
    print(calcRatio(DHL), "star")
```
