### **2. Compare words with the positive, negative and neutral English words using a String-Matching algorithm.**

```python
def calcRatio(arr):  # positive ratio out of 5(put array)
    total = 0
    posiNum = 0
    negNum = 0
    for i in range(len(arr)):
        posiNum += len(arr[i].positive)
        negNum += len(arr[i].negative)

    total = posiNum + negNum

    star = posiNum * 1.0 / total * 5
    return star

ratio = []
if __name__ == '__main__':
    print("CITYLINK")
    CityLink = []
    for i in range(len(citylink)):
        CityLink.append(Website.Website(citylink[i]))
        CityLink[i].read()
        CityLink[i].PosiNeg()
        CityLink[i].Print()
    print(calcRatio(CityLink), "star")
    ratio.append(calcRatio(CityLink))
    print()

    print("POSLAJU")
    Poslaju = []
    for i in range(len(poslaju)):
        Poslaju.append(Website.Website(poslaju[i]))
        Poslaju[i].read()
        Poslaju[i].PosiNeg()
        Poslaju[i].Print()
    print(calcRatio(Poslaju), "star")
    ratio.append(calcRatio(Poslaju))
    print()

    print("GDEX")
    GDEX = []
    for i in range(len(poslaju)):
        GDEX.append(Website.Website(gdex[i]))
        GDEX[i].read()
        GDEX[i].PosiNeg()
        GDEX[i].Print()
    print(calcRatio(GDEX), "star")
    ratio.append(calcRatio(GDEX))
    print()

    print("J&T")
    JT = []
    for i in range(len(poslaju)):
        JT.append(Website.Website(jt[i]))
        JT[i].read()
        JT[i].PosiNeg()
        JT[i].Print()
    print(calcRatio(JT), "star")
    ratio.append(calcRatio(JT))
    print()

    print("DHL")
    DHL = []
    for i in range(len(poslaju)):
        DHL.append(Website.Website(dhl[i]))
        DHL[i].read()
        DHL[i].PosiNeg()
        DHL[i].Print()
    print(calcRatio(DHL), "star")
    ratio.append(calcRatio(DHL))
```

### **4. Plot histogram graphs of positive and negative words.**


```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

wordCountData = pd.read_csv('wordCount.csv')
df = pd.DataFrame(wordCountData)
display(df)
print()
ct_df = df[df['company'] == "CityLink"]
pj_df = df[df['company'] == "Poslaju"]
gd_df = df[df['company'] == "GDEX"]
jnt_df = df[df['company'] == "J&T"]
dhl_df = df[df['company'] == "DHL"]
dfs = ["CityLink", "Poslaju", "GDEX", "J&T", "DHL"]

def f(Company):
    if Company == 'CityLink':
        df = ct_df
    elif Company == 'Poslaju':
        df = pj_df
    elif Company == 'GDEX':
        df = dhl_df
    elif Company == 'J&T':
        df = jnt_df
    else:
        df = dhl_df
     
    
    fig = go.Figure()
    fig.add_trace(go.Bar(x = ['Article 1', 'Article 2', 'Article 3'],
                         y = df['word_count'], 
                         name = "Word Count"))
    fig.add_trace(go.Bar(x = ['Article 1', 'Article 2', 'Article 3'],
                         y = df['stopword_count'],
                         name = "Stop Word",
                         opacity=0.6))
    fig.update_xaxes(type='category')
    fig.update_layout(barmode="overlay", bargap=0.1)
    fig.show()
    
    fig2 = go.Figure()
    fig2.add_trace(go.Histogram(histfunc="sum", y=df['pov'], x=['Article 1', 'Article 2', 'Article 3'], name="Positive word"))
    fig2.add_trace(go.Histogram(histfunc="sum", y=df['neg'], x=['Article 1', 'Article 2', 'Article 3'], name="Negative word"))
    fig2.show()

if __name__ == "__main__":
    interact(f, Company = dfs)

```
